```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt


# ДИНАМИКА РОССИЙСКИХ АЭРОПОРТОВ ПО ЧИСЛУ ПАССАЖИРАВ
```

```python
names = pd.read_csv(
    'Experimental_data/russian_air_service_CARGO_AND_PARCELS.csv',)
information_2019 = names[names['Year'] == 2019].sort_values('Whole year', ascending=False).head(15)
information_2019
```

```python
plt.barh(information_2019['Airport name'], information_2019['Whole year'])
plt.xlabel('Number of passengers')
plt.ylabel('Airport')
plt.title('Popular airports 2019')
plt.grid()
# САМЫЕ ПОПУЛЯРНЫЕ АЭРОПОРТЫ (ШЕРЕМЕТЬЕВО В РАЗЫ КРУПНЕЕ ВСЕХ)
```

```python
total = information_2019.dropna()
#total.head()
total_moscow_summer = total[[x for x in total.columns if x in ['Whole year','June','July','August']] + ['Airport name']].groupby('Airport name').sum()
print(total_moscow_summer)
```

```python
total_moscow_summer.plot()
# ДИНАМИКА НАСКОЛЬКО БОЛЬШЕ ПАССАЖИРОВ У ШЕРЕМЕТЬЕВО ПАССАЖИРОВ
```

```python
Sheremetyevo  = names[names['Airport name'] == 'Moscow (Sheremetyevo)']
Sheremetyevo_year = Sheremetyevo[['Year'] + ['January'] + ['February'] + ['August'] + ['March'] + ['May'] + ['June'] + ['July'] + ['September'] + ['October'] + ['November'] + ['December']].groupby('Year').sum()
Sheremetyevo_year.plot()
```

```python
Sheremetyevo_year[list(filter(lambda x: x != 'Whole year', Sheremetyevo_year.columns))]\
    .plot(kind='bar', rot=45, stacked=True)


# ДИНАМИКА ПО МЕСЯЦАМ И ВСЕГО ЗА ГОДЫ
# В 2015 КРИЗИС И ПРОВАЛ В ПОЛЕТАХ
```

```python
Sheremetyevo_year[list(filter(lambda x: x == 'January', Sheremetyevo_year.columns))]\
    .plot(kind='area', rot=45, stacked=False)



# ТАКЖЕ ВИДНО ЧТО В КРИЗИС ПАДАЕТ КОЛЛИЧЕСТВО ПОЛЕТОВ В КУРОРТНЫЙ МЕСЯЦ 
```
